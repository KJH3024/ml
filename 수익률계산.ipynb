```python
#Packages used (사용된 패키지)
import yfinance as yf
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import FinanceDataReader as fdr
import matplotlib

#Download price data from Yahoo finance (가격 데이터 다운로드)
p_kakao = fdr.DataReader('293490','2022-06-15', end='2022-07-05')
p_el = fdr.DataReader('066970','2022-06-15')
p_ho = fdr.DataReader('008770','2022-06-15')
p_kospi = fdr.DataReader('KS200','2022-06-15')
p_kosdaq = fdr.DataReader('KQ11','2022-06-15')
```

```python
new_data = pd.DataFrame()

new_data['카카오게임즈'] = p_kakao['Close']
new_data['호텔신라'] = p_ho['Close']
new_data['엘앤에프'] = p_el['Close']
new_data['코스피200'] = p_kospi['Close']
new_data['코스닥'] = p_kosdaq['Close']
```

```python
new_data
```

```python
df = pd.DataFrame()
```

```python
df['카카오게임즈'] = ((p_kakao["Close"] - p_kakao['Close'].shift())/ p_kakao['Close'].shift() * 100).round(2)
df['호텔신라'] = ((p_ho['Close'] - p_ho['Close'].shift())/ p_ho['Close'].shift() * 100).round(2)
df['엘앤에프'] = ((p_el['Close'] - p_el['Close'].shift())/ p_el['Close'].shift() * 100).round(2)
df
```

```python
matplotlib.rcParams['font.family'] = 'Malgun Gothic'
matplotlib.rcParams['axes.unicode_minus'] = False
plt.figure(figsize=(16,8))
plt.plot(df)
plt.title('일별 수익률', fontsize = 15)
plt.ylabel('수익률(%)', fontsize = 13)
plt.xticks(fontsize = 15)
plt.yticks(fontsize = 15)
plt.legend(['카카오게임즈','호텔신라','엘앤에프'], fontsize = 15)
plt.show()
```
